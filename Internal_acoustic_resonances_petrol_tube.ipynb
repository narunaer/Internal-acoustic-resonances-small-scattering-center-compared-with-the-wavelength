```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import eval_legendre, lpmv, spherical_jn, spherical_yn
from scipy import integrate
from scipy.integrate import quad
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.colors import LogNorm
from scipy.special import eval_legendre
from scipy.misc import derivative
from matplotlib.ticker import ScalarFormatter
from matplotlib.colors import LinearSegmentedColormap
from scipy.signal import find_peaks
from matplotlib.ticker import LogLocator, LogFormatter
import matplotlib.ticker as ticker
import scipy.special as sp


dpi= 300
px = 1e0/dpi

plt.rcParams.update({
    "text.usetex"   : False,
    "figure.dpi"    : dpi,
    #"text.latex.preamble" : r"\usepackage{lmodern}",
    "font.family" : "sans-serif",
#    "figure.figsize": (800*px,600*px)
})
plt.rcParams['text.usetex'] = True

B = 1
f = 1
c1 = 1250e0
c0 = 1522e0


rho1 = 775
rho0 = 1024e0

m = c0 / c1

k0 = 2*np.pi*f/c0
k1 = 2*np.pi*f/c1
x0_max = 10
a = 0.1e0
x0_range = np.arange(0,0.2,1e-5)
q= 2*f/(c1*rho1*B**2)
mt = (rho0/rho1)*m
w = 2*np.pi*f
```

```python
J = lambda n, x: sp.jn(abs(n), x) * ((-1e0) ** n)  # Corrects for negative n using (-1)^n
H = lambda n, x: sp.jn(abs(n), x) + 1e0j * sp.yn(abs(n), x)  # For Hankel functions, no need to treat negative n separately
dJ = lambda n, x: sp.jvp(abs(n), x) * ((-1e0) ** n)  # Derivative for J
dH = lambda n, x: sp.jvp(abs(n), x) + 1e0j * sp.yvp(abs(n), x)  # Derivative for H

def b_l(l, m, mt, x0, x1):
    J = lambda n, x: sp.jn(abs(n), x) * ((-1e0) ** n)  # Corrects for negative n using (-1)^n
    H = lambda n, x: sp.jn(abs(n), x) + 1e0j * sp.yn(abs(n), x)  # For Hankel functions, no need to treat negative n separately
    dJ = lambda n, x: sp.jvp(abs(n), x) * ((-1e0) ** n)  # Derivative for J
    dH = lambda n, x: sp.jvp(abs(n), x) + 1e0j * sp.yvp(abs(n), x)  # Derivative for H

    tmp = (mt * dJ(l, x1) * H(l, x0)- J(l, x1) * dH(l, x0) )

    if tmp == 0:
        raise ValueError("Denominator in the calculation of bl is zero, which will cause a division by zero error.")

    bl = (dJ(l, x0) * H(l, x0) - J(l, x0) * dH(l, x0)) * (mt / m) / tmp

    return bl

def I_1(l, x1):
    integrand = lambda x: x*J(l-1, x)*J(l+1, x)
    result, _ = quad(integrand, 0, x1)
    return  result


def I_wp(l, x1):
    integrand = lambda x: x*(J(l, x)**2)
    result, _ = quad(integrand, 0, x1)
    return  result

def I_wk1(l, x1):
    integrand = lambda x: (J(l, x)**2)/x
    result, _ = quad(integrand, 0, x1)
    return  result
def I_wk2(l, x1):
    integrand = lambda x: (dJ(l, x)**2)*x
    result, _ = quad(integrand, 0, x1)
    return  result
def ell_max(x0):
    return 3 + x0 + 4.05 * x0**(1/3)
```

```python
x0_vector = np.linspace(1e-5, 1, 1000)
x0_vector_ray = np.linspace(1e-5, 0.1, 1000)
phi_vector = np.linspace(0, np.pi, 1000)

velocity_ray = np.zeros((len(x0_vector), len(phi_vector)), dtype=complex)
velocity_phi_ray = np.zeros((len(x0_vector), len(phi_vector)), dtype=complex)
pressure_ray = np.zeros((len(x0_vector), len(phi_vector)), dtype=complex)

for i in range(len(x0_vector_ray)):
    r_ray = x0_vector_ray[i]
    x1_ray = m * x0_vector_ray[i]

    for j in range(len(phi_vector)):
        phi = phi_vector[j]
        sum_result_vray = 0
        sum_result_vphi_ray = 0
        sum_result_pray = 0
        b0 = b_l(0, m, mt, r_ray, x1_ray)
        sum_result_vray +=  k1 * b0* dJ(0, x1_ray)
        sum_result_vphi_ray +=  0
        sum_result_pray += b0 * J(0, x1_ray)
        velocity_ray[i, j] = B * sum_result_vray
        velocity_phi_ray[i, j] = B * sum_result_vphi_ray
        pressure_ray[i, j] = - rho1 * 1j * c1 * k1 * B * sum_result_pray

velocity_r = np.zeros((len(x0_vector), len(phi_vector)), dtype=complex)
velocity_phi = np.zeros((len(x0_vector), len(phi_vector)), dtype=complex)
pressure = np.zeros((len(x0_vector), len(phi_vector)), dtype=complex)

l_values = range(0, int(ell_max(x0_max)))

for i in range(len(x0_vector)):
    r = x0_vector[i]
    x1_vector = m * x0_vector[i]
    for j in range(len(phi_vector)):
        phi = phi_vector[j]
        sum_result_vr = 0
        sum_result_vphi = 0
        sum_result_p = 0
        for l in l_values:
            bl = b_l(l, m, mt, r, x1_vector)
            epsilon_l = 1 if l == 0 else 4

            sum_result_vr += bl * epsilon_l * (1j ** l) *dJ(l, x1_vector)*np.cos(l*phi)

            sum_result_vphi += - bl * l * epsilon_l * (1j ** l) * J(l, x1_vector)* np.sin(l*phi) * (1/ x1_vector)

            # Pressure component
            sum_result_p += bl * epsilon_l * (1j ** l) * J(l, x1_vector) * np.cos(l*phi)

        velocity_r[i, j] = k1 * B * sum_result_vr
        velocity_phi[i, j] = k1 * B * sum_result_vphi
        pressure[i, j] = - c1 * k1 * 1j * rho1 * B * sum_result_p
```
